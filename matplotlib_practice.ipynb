```python
import matplotlib.pyplot as plt
import statistics
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# plt.figure(figsize=(500,350))
plt.plot(months, savings, label='Savings')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Saving Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
z_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 0, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.figure(figsize=(5,2.5))
plt.plot(x_axis, y_axis, label='Boston', color='green', marker='*', linewidth=1)
# plt.figure(figsize=(20,10))
plt.xlabel('Month')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.ylim(0,45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots(figsize=(12,3))
ax.plot(x_axis, y_axis, label='Boston', color='green', linewidth=0.25, marker='^')
ax.set_xlabel('Month')
ax.set_ylabel('Fare ($)')
ax.legend()
ax.grid()
ax.set_ylim(0,45)
ax.set_title('PyBer Fare by Month')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
ax = plt.axes()
ax.plot(x_axis, z_axis)
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.bar(x_axis,y_axis, width=0.5, color='green', label='Boston')
plt.xlabel("Date")
plt.ylabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.legend()
# plt.grid()
plt.show()
```

```python
plt.barh(x_axis,y_axis, color='magenta', label='Boston')
plt.xlabel("Date")
plt.ylabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.legend()
plt.gca().invert_yaxis()
# plt.grid()
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(6,6))
ax.bar(x_axis, y_axis, width=0.7, color='k', label='New York')
ax.legend()
plt.show()
```

```python
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis, label='Chicago',color='cyan')
plt.gca().invert_yaxis()
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.legend()
ax.set_title('PyBer Fares by Month')
plt.show()
```
